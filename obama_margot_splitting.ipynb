```python
import os
import shutil

# Define the directory to search, the target directory, and the string to search for
search_dir = 'data/SEPARATED/IMAGES/TAYLOR/TRAINING/FAKE_IMG'
target_dir = 'data/SEPARATED/IMAGES/TAYLOR/TEST/FAKE'
search_string = '-to-taylor'

# Create the target directory if it doesn't exist
if not os.path.exists(target_dir):
    os.makedirs(target_dir)

# Walk through the directory
for dirpath, dirnames, files in os.walk(search_dir):
    for file in files:
        # Check if the search string is in the file name
        if search_string.lower() in file.lower():
            # Construct full file paths
            src_file_path = os.path.join(dirpath, file)
            dst_file_path = os.path.join(target_dir, file)
            
            # Copy the file to the new directory
            shutil.copy2(src_file_path, dst_file_path)
            
            # Delete the file from the original directory
            os.remove(src_file_path)
            print(f'File {file} has been copied and deleted.')

print('Operation completed.')
```
